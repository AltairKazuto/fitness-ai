```python
from ultralytics import YOLO

model = YOLO('yolo11n-pose.pt')

#train with default parameters
#add patience to end as model converges
#save every 10 epochs to avoid loss of progress

model.train(
    data='/content/Workout_Pose-11/data.yaml',
    epochs=120,
    imgsz=640,
    name='yoga_pose_classifier',
    patience=20,
    save=True,
    save_period=10
)
```
